# Linear Regression Algorithm

```python
#  regression - continous value at target point
#  linear regression- a data analysis technique that predicts the value of unknown data by using another related and known data value
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
area=np.array([700,800,900,100,1200,1500])
price=np.array([25,28,35,38,39,40])
```

```python
plt.scatter(area,price)
plt.show()
```

```python
# line equaton-->  y=mx+c  (where m--> slope,   c--> intercept) formula= y2-y1/x2-x1.
# y^ --> predicted y (where  y-y^)=error/residual
```

```python
#find y in y=mx+c

x=[1,2,3,4]
m=0.5
c=4
y=[m*x+c for x in x]
print(y)
```

```python
plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
# in linear regression cost function is MSE(mean square error)
# MSE= summation of (y-y^)^2/n (where n is total no. of data points or the records in table)
```

```python
# let "z" is predicted y(predicted price) value here....in above question of price and are
```

```python
area=np.array([700,800,900,100,1200,1500])
price=np.array([25,28,35,38,39,40])
m=1
c=2
z=[m*i+c for i in area]
print(z)
```

```python
plt.scatter(area,price)
plt.plot(area,z)
plt.show()
```

```python
#  summation of (y-y^)^2 is SSE(Sum Squared Error)
#  sqrt of summation of (y-y^)^2/n is RMSE(Root Mean Squared Error)
```

```python
# y^=mx+c (where m--> theta0 and c--> theta1; theta0 and theta1 are coefficients)
```

```python
area=np.array([700, 800, 900, 1000, 1200, 1500])
price=np.array[25, 28, 35, 38, 39, 40]
```

```python
x_mean=np.mean(area)    #area mean
x_mean
```

```python
y_mean=np.mean(price)  # price mean
y_mean
```

```python
numerat = np.sum((area-x_mean) *(price-y_mean))
numerat
```

```python
denom = np.sum((area-x_mean) **2)
denom
```

```python
tithe1=numerat/denom
tithe1
```

```python
titha0 = (y_mean - (tithe1*x_mean))
titha0
```

```python
pred_price = ((tithe1 *area) + titha0)
pred_price               # predicted  price
```

```python
plt.scatter(area, price)
plt.plot(area, pred_price)
plt.show()
```

```python
plt.hist(price-pred_price) # (y-y^) is called as Residual or Error
```

```python
MSE=np.sum((price-pred_price)**2)/6
MSE
```

```python
plt.hist(MSE)
plt.show()
```
